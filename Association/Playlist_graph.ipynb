```python
df=pd.read_csv("C:/Users/I068117/UT_Machine Learning/Association/playlists.csv")
df.head()
```

```python
df=df.groupby('user')['artist'].unique().apply(list).reset_index(drop=True)
```

```python
df
```

```python
records=[df[i] for i in range(0,len(df)-1)]
records
```

```python
!pip install mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
```

```python
## Remove duplicates ("de-dupe")
#playlists = playlists.apply(pd.unique)

## Cast this variable as a special arules "transactions" class.
playtrans = TransactionEncoder()
playtrans_ary = playtrans.fit(records).transform(records)
playtrans_df = pd.DataFrame(playtrans_ary, columns=playtrans.columns_)
playtrans_df
```

```python
# Compute the support
support = playtrans_df.mean()
support = pd.DataFrame(support, columns=['support']).sort_values('support',ascending=False)

# Print the support
support.head()
```

```python
support.describe()
```

```python

# Now run the 'apriori' algorithm
# Look at rules with support > .005 & confidence >.1 & length (# artists) <= 4
musicrules = apriori(playtrans_df, min_support=0.03, use_colnames=True)
musicrules = association_rules(musicrules, metric="confidence", min_threshold=0.1)

# Look at the output... so many rules!
#print(musicrules)
```

```python
musicrules.sort_values(by=['lift'],ascending=False)
```

```python
musicrules[musicrules['itemsets'].apply(lambda x: len(x) > 1)].sort_values(by=['support'],ascending=False)
```
