```python
!pip install mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
```

```python
record=[]
file = open("C:/Users/I068117/UT_Machine Learning/Association/groceries.txt", "r")
while True:
    content=file.readline()
    record.append(set(content.strip().split(',')))
    if not content:
        break
    #print(content)
file.close()
```

```python
## Cast this variable as a special arules "transactions" class.
playtrans = TransactionEncoder()
playtrans_ary = playtrans.fit(record).transform(record)
playtrans_df = pd.DataFrame(playtrans_ary, columns=playtrans.columns_)
playtrans_df
```

```python
# Compute the support
support = playtrans_df.mean()
support = pd.DataFrame(support, columns=['support']).sort_values('support',ascending=False)
# Print the support
support.head()
```

```python
support.describe()
```

```python
musicrules = apriori(playtrans_df, min_support=0.01, use_colnames=True)
musicrules = association_rules(musicrules, metric="confidence", min_threshold=0.1)
```

```python
musicrules.sort_values(by=['lift'],ascending=False)
```

```python
filtered=musicrules[musicrules['lift']>2]
filtered[filtered['consequents'].apply(lambda x: 'whole milk' in x)]
```

```python
import networkx as nx

# Assuming 'musicrules' is your DataFrame with association rules

# Filter rules based on lift threshold
filtered_rules = musicrules[musicrules['lift'] > 2]

# Create a directed graph
playlists_graph = nx.DiGraph()

# Add edges to the graph
for _, rule in filtered_rules.iterrows():
    for lhs_item in rule['antecedents']:
        playlists_graph.add_edge(lhs_item, rule['consequents'])

# Write the graph to a GraphML file
nx.write_graphml(playlists_graph, 'groceries_python.graphml')
```
